```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

print(df)

# Bar chart
plt.bar(df["Month"], df["Sales"])
plt.title("Monthly Sales")
plt.savefig("bar.png")
plt.show()

# Pie chart
plt.pie(df["Sales"], labels=df["Month"], autopct="%1.1f%%")
plt.title("Sales Distribution")
plt.savefig("pie.png")
plt.show()
```
